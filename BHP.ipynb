```python
def convert(x):
    r = x.split(" - ")
    if len(r) == 2:
        return ((float(r[0]) + float(r[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
def remove_pps_outlier(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby ('location'):
#        print(key)
#        print(subdf)
       m = np.mean(subdf.price_per_ft)
       s = np.std(subdf.price_per_ft)
#       print(m)
#       print(s)
       reduced_df = subdf[(subdf.price_per_ft >(m-s)) & (subdf.price_per_ft < (m+s))]
#       print(reduced_df)
       df_out = pd.concat([df_out, reduced_df], ignore_index = True)
    return df_out
```
